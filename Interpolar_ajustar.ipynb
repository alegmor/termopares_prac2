```python
import pandas as pd
import numpy as np

%config Completer.use_jedi = False
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
#importar archivo de datos
df = pd.read_csv('Tabla_termopar_tipoT1.csv',decimal=',')
df
```

```python
#editar archivo de datos
datos = pd.read_csv('datos.csv',decimal=',')
datos.dtypes
datos1 = pd.DataFrame(datos['102 (Vdc)'])
datos1.columns
datos1.rename(columns = {'102 (Vdc)':'V'}, inplace = True)
NaN = np.nan
datos1["T"] = NaN
datos1 = datos1[["T","V"]]
datos1
```

```python
df=df.append(datos1,ignore_index=True)
```

```python
print(df)
```

```python
df = df.sort_values(by='V')
df
```

```python
dfi = df.interpolate()
```

```python
import matplotlib.pyplot as plt

plt.plot(dfi["T"],dfi["V"])
plt.xlabel("Temperatura [$^\circ C$]")
plt.ylabel("Voltaje [$mV$]")
plt.show()
```

```python
dfir = dfi.sort_index()
dfir
```

```python
data = dfir.iloc[31:]
plt.plot(data["T"],data["V"],'*')
plt.xlabel("Temperatura [$^\circ C$]")
plt.ylabel("Voltaje [$mV$]")
plt.show()
```

```python
s1,s0 = np.polyfit(data['T'],data.V,1)
plt.plot(data['T'],data.V,'*')
plt.plot(data['T'],s1*data['T']+s0,color='black',label='v(T)={}T{}'.format(round(s1,4),round(s0,4)))
plt.xlabel("Temperatura [$^\circ C$]")
plt.ylabel("Voltaje [$mV$]")
plt.legend()
plt.show()
```
